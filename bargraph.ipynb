```python
import matplotlib.pyplot as plt

categories = ['Rent', 'Groceries', 'Utilities', 'Entertainment', 'Transportation']
expenses = [1200, 300, 150, 200, 100]

# Plotting the bar chart
plt.figure(figsize=(8,4))
plt.bar(categories, expenses, color='skyblue')

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Monthly Expenses ($)')
plt.title('Monthly Expenses in Different Spending Categories')

# Display the plot
plt.show()
```

i) The highest expense is in the 'Rent' category, indicating a significant allocation of the monthly budget to housing costs.

 ii) 'Groceries' and 'Entertainment' show moderate expenses, suggesting controlled spending on essential items.

 iii) 'Utilities' and 'Transportation' have lower expenses, indicating relatively limited discretionary spending in these areas.
