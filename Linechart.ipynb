```python
import matplotlib.pyplot as plt
```

```python
temperature_data = [32,31,30,28,32,33,34,36,36,36,36,36,36,36,34,34,33,33,33,33,33,32,32,31,32,33,31,31,32,32,33]

days = list(range(1, 32))

# Plotting the daily temperature changes
plt.figure(figsize=(6, 2))
plt.plot(days, temperature_data, marker='o', linestyle='-', color='b')

# Adding labels and title
plt.xlabel('Days of the month')
plt.ylabel('Temperature (°C)')
plt.title('Daily Temperature Changes(October)')

# Display the plot
plt.show()
```

The temperatures seem to be within a certain range, with a maximum around 36°C and a minimum around 28°C
